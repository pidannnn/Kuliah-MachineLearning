```python
import pandas as pd

# baca data
df = pd.read_csv("../data/day.csv")

# mengacak data
df = df.sample(frac = 1, random_state = 42).reset_index(drop = True)

# hitung jumlah data
total = len(df)
train_size = int(0.8 * total) # 80% pertama untuk training + validasi
val_size = int(0.1 * train_size) # 10% dari training untuk validasi

train_data = df[:train_size]
test_data = df[train_size:]

val_data = train_data[:val_size]
train_data = train_data[val_size:]
```

```python
print("Total data     :", total)
print("Train data     :", len(train_data))
print("Validation data:", len(val_data))
print("Test data      :", len(test_data))
```

```python
# data training
print("Data Training")
print("Jumlah data:", len(train_data))
train_data.head()
```

```python
# data validasi
print("\nData Validasi")
print("Jumlah data:", len(val_data))
val_data.head()
```

```python
# data testing
print("\nData Testing")
print("Jumlah data:", len(test_data))
test_data.head()
```
